```python
embedings_test = tr.get_embedings()
Ytrue = tr.get_y_true()
```

```python
cosine_similarity = tf.keras.metrics.CosineSimilarity()
counter = 0
epoche = 0 

epochs_dic = {}
answer_list = []

true_dic = {}
true_list =[]


def cos_sim(a,p,n): 

    a_mag = tf.math.sqrt(tf.math.reduce_sum(tf.multiply(a,a), axis = -1))        # Magnitude of the Anchor Vector
    p_mag = tf.math.sqrt(tf.math.reduce_sum(tf.multiply(p,p), axis = -1))        # Magnitude of the Postive Vector
    n_mag = tf.math.sqrt(tf.math.reduce_sum(tf.multiply(n,n), axis = -1))        # Magnitude of the Negativ Vector 

    ap = tf.math.reduce_sum(tf.multiply(a,p))                                    # Scalar Product of Anchor and Positive
    an = tf.math.reduce_sum(tf.multiply(a,n))                                    # Scalar Product of Anchor and Negative

    cos_ap = ap / tf.multiply(a_mag, p_mag)                                      # Cosinus between Anchor and Postive 
    cos_an = an / tf.multiply(a_mag, n_mag)                                      # Cosinus between Anchor and Negative 

    if cos_ap < cos_an:     
        return 1                                                      # Retunr 1, becuase the Angle between Anchor and Positive is smaller than Anchor and Negative  
    else: 
        return 0                                                      # Return 0, becuase the Angle between Anchor and Negative is smaller than Anchor and Positive 

        
for j in range(len(embedings_test)): 
    anchor, pos , neg = embedings_test[j]
    y_t = Ytrue[j]
    for i in range(2):
        cos_sim(anchor[i,:], pos[i,:], pos[i,:])
        pos_sim = cosine_similarity(anchor[i,:], pos[i,:])
        neg_sim = cosine_similarity(anchor[i,:], neg[i,:])
        true_list.append(y_t.numpy()[i][0])
        if pos_sim < neg_sim: 
            answer_list.append(1.)
        else: 
            answer_list.append(0.)

    if(counter%10):
        epochs_dic["Epoch"+str(epoche)] = answer_list
        true_dic["Epoch"+str(epoche)] = true_list
    
    counter += 1

    if counter == 11: 
            epoche += 1 
            counter = 0 
            answer_list = []  
            true_list = []
acc_list = []

for element in epochs_dic: 
    epoch_pred = epochs_dic[element]
    epoch_true = true_dic[element]

    sum = 0
    for a in range(len(epoch_pred)): 
        if epoch_pred[a] == epoch_true[a]: 
            sum += 1
    acc = sum / len(epoch_pred)
    acc_list.append(acc)


print("Max. Acc:")
print(max(acc_list))
index = acc_list.index(max(acc_list))
epoch_max = "Epoch"+str(index)
print(epoch_max)
print("Prediction:")
print(epochs_dic[epoch_max])
print("True:")
print(true_dic[epoch_max])


    
```
